### Read CSV File

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as pyt 
data=pd.read_csv("company_total_sales.csv")
data.head(5)


```

### Line Plot

```python
sns.set_style("darkgrid")
sns.lineplot(x="Months",y="Total Sales",data=data)
pyt.title("Trends of Company Sales Over a Year")
pyt.ylabel("Total Per month Sales")
pyt.xlabel("Months")
pyt.show()
```
