### Read csv File 

```python
import numpy as np
import pandas as pd 
df=pd.read_csv("student_scores.csv") # Read csv File
df.head(10)
```

### Calculate average score of each student 

```python
df["Average Score"]=df[["Math","Science","English"]].mean(axis=1)
df.head(10)
```

### Student With Highest Average 

```python
print("Student Having highest Average score is given below:")
max=df["Average Score"].max() #Find maximum Value from Average Score 
mss=df[df["Average Score"]==max] #Filter row by Average score is equal to Maximum
std_name=mss["Student Name"]#Retrive Student name form filtered row 
print(std_name)#print Student name having highest score 
```
